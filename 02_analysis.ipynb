```python
# Load dataset
df = pd.read_csv("cleaned_dataset (1).csv", encoding="latin1")

# 1. Descriptive stats for numerical columns
print("ðŸ”¸ Numerical Columns Statistics:")
print(df.describe())

# 2. Explore categorical columns
print("\nðŸ”¹ Unique Country Codes:", df["Country Code"].nunique())
print("Top Countries:\n", df["Country Code"].value_counts().head(5))

print("\nðŸ”¹ Unique Cities:", df["City"].nunique())
print("Top Cities:\n", df["City"].value_counts().head(10))

print("\nðŸ”¹ Unique Cuisines:", df["Cuisines"].nunique())
print("Top Cuisines:\n", df["Cuisines"].value_counts().head(10))

# 3. Grouping by insights
top_cities = df["City"].value_counts().head(5)
top_cuisines = df["Cuisines"].value_counts().head(5)

# Optional: Export summaries
summary = {
    "Top Cities by Count": top_cities,
    "Top Cuisines by Count": top_cuisines,
    "Numerical Summary": df.describe()
}
plt.figure(figsize=(10, 4))
sns.countplot(y=df["City"], order=df["City"].value_counts().head(10).index, palette="coolwarm")
plt.title("Top 10 Cities with Most Restaurants")
plt.xlabel("Count")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```
