```python
import pandas as pd

# Load cleaned dataset
df = pd.read_csv("cleaned_dataset (1).csv", encoding="latin1")

# ──────────────────────────────────────
# 1. Extract Length-Based Features
# ──────────────────────────────────────

# Length of restaurant name
df["Name_len"] = df["Restaurant Name"].astype(str).apply(len)

# Length of address (if column exists)
if "Address" in df.columns:
    df["Address_len"] = df["Address"].astype(str).apply(len)
else:
    print("⚠️ 'Address' column not found, skipping address length feature.")

# ──────────────────────────────────────
# 2. Encode Categorical Binary Columns
# ──────────────────────────────────────

# Encode "Has Table booking" column
df["Has_Table"] = df["Has Table booking"].map({"Yes": 1, "No": 0})

# Encode "Has Online delivery" column
df["Has_Delivery"] = df["Has Online delivery"].map({"Yes": 1, "No": 0})

# Check result
df[["Restaurant Name", "Name_len", "Has Table booking", "Has_Table", "Has Online delivery", "Has_Delivery"]].head()
```
